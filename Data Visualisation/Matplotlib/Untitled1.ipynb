```python
import seaborn as sbn
```

```python
%matplotlib inline
```

```python
tips= sbn.load_dataset('tips')
```

```python
tips.head()
```

```python
sbn.distplot(tips['total_bill'],kde=False)
```

```python
sbn.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

```python
sbn.pairplot(tips,hue='sex')
```
